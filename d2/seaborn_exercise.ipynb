## Seaborn Assignment

```python
# import pandas
import pandas as pd

# import seaborn
import seaborn as sns

# import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('https://gist.githubusercontent.com/armgilles/194bcff35001e7eb53a2a8b441e8b2c6/raw/92200bc0a673d5ce2110aaad4544ed6c4010f687/pokemon.csv',sep=',')
df.drop('#',axis=1,inplace=True)
```

### Task 1: Plot a regression line between attack and defense.

```python
# a regression line with a confidence inetrval plot
sns.lmplot(x='Attack', y='Defense', data = df)
plt.show()
```

### Task 2: Plot a regression line between attack and defense while displaying a distinction between the different generations. 

```python
# a regression line with a confidence inetrval plot
sns.lmplot(x='Attack', y='Defense', hue ='Generation' ,data = df)
```

### Task 3: Make a boxplot for each feature of dataframe (without Generation and Legendary features) in one plot.

```python
# Name	Type 1	Type 2	Total	HP	Attack	Defense	Sp. Atk	Sp. Def	Speed	Generation	Legendary
# create a grid with 1 row 2 columns


# the first plot
boxplot = df.boxplot(column=['HP', 'Attack','Defense','Sp. Atk','Sp. Def','Speed'])
```

### Task 4: Plot a correlation heatmap of the pokemon dataframe (not including the features Generation and Legendary).

```python
import numpy as np
newdf = df[df.columns.difference(['Generation','Legendary','Name','Type 1','Type 2'])]

mask = np.zeros_like(newdf)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(7, 5))
    ax = sns.heatmap(newdf, mask=mask, vmax=100, square=True)
```

```python
newdf = df[df.columns.difference(['Generation','Legendary'])]
plt.figure(figsize=(16, 6))
heatmap = sns.heatmap(newdf.corr(), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':18}, pad=12);
# save heatmap as .png file
# dpi - sets the resolution of the saved image in dots/inches
# bbox_inches - when set to 'tight' - does not allow the labels to be cropped
plt.savefig('heatmap.png', dpi=300, bbox_inches='tight')
```
